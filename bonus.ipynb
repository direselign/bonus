## Q#1 For the attributes ‘Primary type’ and ‘Location description’, output the list of each value and the corresponding frequency count, sorted in descending order of the count, respectively.

#### group the data using Primary Type and sort

```python
primaryType = df.groupBy('Primary Type').count().orderBy('count', ascending=False).cache()
primaryType.show(primaryType.count()) 
primaryType
```

#### group the data using Location Description and sort in descending order

```python
locationDescription = df.groupBy('Location Description').count().orderBy('count', ascending=False).cache()
locationDescription.show(locationDescription.count())
```

## Q#2 Output the most frequently occurred ‘Primary type‘ for each possible value of ‘Location description’, sorted in descending order of the frequency count.

```python
primaryTypeOfLocationDescription = df.cube('Primary Type', 'Location Description').count()\
                                          .orderBy('Primary Type', 'count', 'Location Description', ascending=False)\
                                          .cache()
primaryTypeOfLocationDescription.show(primaryTypeOfLocationDescription.count())
```
